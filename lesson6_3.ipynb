```python
import requests
url = 'https://data.ntpc.gov.tw/api/datasets/71cd1490-a2df-4198-bef1-318479775e8a/csv?size=100'

#下載資料
response = requests.get(url)
if response.ok:
    print(response.text)
```

```python
#沒有存檔csv,直接把文字轉為字串 運用StringIO
from io import StringIO
import csv
file = StringIO(response.text,newline='')
csv_reader = csv.DictReader(file)
csv_list = list(csv_reader)
for item_dict in csv_list:
    print(item_dict)
```

```python
#用DataFrame取代for in
import pandas as pd
df = pd.DataFrame(data=csv_list)
df
```

```python
#使用DataFrame的drop刪除欄位
import pandas as pd
df = pd.DataFrame(data=csv_list)
df.drop(columns=['sareaen','snaen','aren','lat','lng'])
```

```python
#使用DataFrame的rename更改欄位名稱
import pandas as pd
df = pd.DataFrame(data=csv_list)
df1 = df.drop(columns=['sareaen','snaen','aren','lat','lng'])
df2 = df1.rename(columns={'\ufeff"sno"':"編號",
                    'sna':'名稱',
                    'tot':'車輛數',
                    'sbi':'可借',
                    'sarea':'行政區',
                    'mday':'日期',
                    'ar':'地址',
                    'bemp':'可還',
                    'act':'狀態'})
```
