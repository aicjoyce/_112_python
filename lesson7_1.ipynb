```python
import requests
url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'
response = requests.request('GET',url)
if response.status_code == 200 :
    print("連線成功")
    all_data = response.json()
    print(type(all_data))
else:
    print(f'連線失敗:{response.status_code}')
```

```python
import pandas as pd
dataFrame = pd.DataFrame(data = all_data)

dataFrame
```

```python
#運用info確認筆數是否正確及資料的類型
import pandas as pd
dataFrame = pd.DataFrame(data = all_data)

dataFrame.info()
```

```python
import pandas as pd
dataFrame = pd.DataFrame(data = all_data,columns =['sna','tot','sbi','sarea','mday','ar','bemp','act'])

dataFrame
```

```python
#選取欄位
dataFrame[['sna','tot','sbi']]
```

```python
#選取列(slice)
dataFrame[67:72]
```

```python
#同時選取欄與列  loc選擇具有特定標籤的行和列  iloc選擇特定整數位置的行和列
#loc[index,column],iloc[index,column]
dataFrame.loc[100:120,['sna','sarea']]
```

```python
#同時選取欄與列
#loc[index,column],iloc[index,column]
dataFrame.iloc[100:120,[0,3]]
```

```python
#搜尋欄位值內的資料
#1先抓出Series

#bool Series
mask = dataFrame['sbi'] <= 3
dataFrame[mask]
```

```python
#搜尋欄位值內的資料
#1先抓出Series

#bool Series
mask = dataFrame['sbi'] <= 3
mask_dataFrame = dataFrame[mask]
mask_dataFrame.to_csv('可借小於3的站點.csv')
```
